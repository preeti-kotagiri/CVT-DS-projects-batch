```python
#import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {"name":['preeti',"varsha",'sai'],
       "loc":["adb","hyd","hyd"],
       "id":[12,25,3],
       "marks":[90,85,65]}
df = pd.DataFrame(data)
df
```

```python
x = np.arange(1,11)
y = np.arange(1,11)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("This is line plot")
plt.plot(x,y,color = 'cyan',linewidth = 5,linestyle = "-.")
plt.show()
```

```python
x = np.arange(1,11)
y = np.arange(1,11)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("This is line plot")
plt.plot(x,y,"c<")
plt.show()
```

```python
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("This is line plot")
plt.plot(df['id'],df['marks'])
plt.show()
```

### Bar graph

```python
plt.xlabel("Id")
plt.ylabel("marks")
plt.xlim(1,30)
plt.title("This is bar plot")
plt.bar(df['id'],df['marks'],width=1,color='g')
```

```python
plt.xlabel("Id")
plt.ylabel("marks")
plt.ylim(15,40)
plt.title("This is bar plot")
plt.bar(['c','c++','python'],[20,30,40],width=0.5,color='g')
```
